```python
# converting the datatime datatype
date['date']=pd.to_datetime(date['date'])
```

```python
date['date_year']=date['date'].dt.year

date.head()
```

# Extract month

```python
date['date_month']=date['date'].dt.month
```

```python
date.head()
```

```python
date['date_month_name']=date['date'].dt.month_name()
```

```python
date.head()
```

# Extract days

```python
date['date_day']=date['date'].dt.day
```

```python
date.head()
```

# Day of week

```python
# week ka konsa day he
date['date_of_week']=date['date'].dt.dayofweek
```

```python
date.head()
```

```python
# day of the week name
date["date_week_name"]=date['date'].dt.day_name()

date=date.drop(columns=['product_id','city_id','orders']).head()
```

```python
date.head()
```

```python
# is weekend?
date['date_is_weekend']=np.where(date['date_week_name'].isin(['Sunday','Saturday']),1,0)
```
