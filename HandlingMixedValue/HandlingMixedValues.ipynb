```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("titanic.csv")
df.head()
# number column is representing the counting of traveling family member
```

```python
# maximum six family is traviling & minimum 1(A)
df['number'].unique()
```

```python
fig=df['number'].value_counts().plot.bar()
fig.set_title('passenger travelling with')
```

```python
# extract numerical part
df['number_numerical']=pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
# extract category part
df['numerical_categorial']=np.where(df['number_numerical'].isnull(),df['number'],np.nan)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_num']=df['Cabin'].str.extract('(\d+)') #capture numerical part (regular expression )
df['cabin_cat']=df['Cabin'].str[0] #capture the first latter
df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
